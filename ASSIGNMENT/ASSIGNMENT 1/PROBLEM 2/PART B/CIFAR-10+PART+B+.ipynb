```python
import keras 
import numpy as np

from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Dense , Dropout , Flatten , Conv2D , MaxPooling2D,Activation
from keras.models import Sequential
from keras.utils import np_utils
import config

(x_train , y_train) , (x_test , y_test) = cifar10.load_data()

x_train  = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

y_train = np_utils.to_categorical(y_train , 10)
y_test = np_utils.to_categorical(y_test , 10)
```

```python
model_architecture = input('Enter a number between 1 and 8')
if model_architecture == '1':
    model_architecture = config.Hyperparameters_1
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture == '2':
    model_architecture = config.Hyperparameters_2
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture == '3':
    model_architecture = config.Hyperparameters_3
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture == '4':
    model_architecture = config.Hyperparameters_4
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture == '5':
    model_architecture = config.Hyperparameters_5
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture =='6' :
    model_architecture = config.Hyperparameters_6
    model_architecture = model_architecture.__dict__
    print(model_architecture)
elif model_architecture == '7':
    model_architecture = config.Hyperparameters_7
    model_architecture = model_architecture.__dict__
    print(model_architecture)
else :
    model_architecture = config.Hyperparameters_8
    model_architecture = model_architecture.__dict__
    print(model_architecture)
```

```python
dictlist = []
for key, value in model_architecture.items():        
    temp = [key,value]
    dictlist.append(temp)
#print(dictlist)

hidden_layers = dictlist[1]
hidden_layers = hidden_layers[1]

batch_size = dictlist[2]
batch_size = batch_size[1]

num_classes = dictlist[3]
num_classes = num_classes[1]

epochs = dictlist[4]
epochs = epochs[1]

nodes_1 = dictlist[5]
nodes_1 = nodes_1[1]

nodes_2 = dictlist[6]
nodes_2 = nodes_2[1]

filters = dictlist[7]
filters = filters[1]

dropout_1 = dictlist[8]
dropout_1 = dropout_1[1]

dropout_2 = dictlist[9]
dropout_2 = dropout_2[1]

dropout_3 = dictlist[10]
dropout_3 = dropout_3[1]

data_augmentation = dictlist[11]
data_augmentation = data_augmentation[1]

max_pooling =  dictlist[12]
max_pooling =  max_pooling[1]

activation_1 = dictlist[13]
activation_1 = activation_1[1]

activation_2 = dictlist[14]
activation_2 = activation_2[1]

activation_final_layer = dictlist[15]
activation_final_layer = activation_final_layer[1]

dense  = dictlist[16]
dense  = dense [1]

optimizer = dictlist[17]
optimizer = optimizer[1]

num_predictions = dictlist[18]
num_predictions = num_predictions[1]
```

```python
model = Sequential()

model.add(Conv2D(nodes_1, (filters, filters), padding='same',
                 input_shape=x_train.shape[1:]))
model.add(Activation(activation_1))
model.add(Conv2D(nodes_1, (filters, filters)))
model.add(Activation(activation_2))
model.add(MaxPooling2D(pool_size=(max_pooling, max_pooling)))
model.add(Dropout(dropout_1))

model.add(Conv2D(nodes_2, (filters, filters), padding='same'))
model.add(Activation(activation_2))
model.add(Conv2D(nodes_2, (filters, filters)))
model.add(Activation(activation_2))
model.add(MaxPooling2D(pool_size=(max_pooling, max_pooling)))
model.add(Dropout(dropout_1))

model.add(Flatten())
model.add(Dense(dense))
model.add(Activation(activation_1))
model.add(Dropout(dropout_2))
model.add(Dense(num_classes))
model.add(Activation(activation_final_layer))

model.compile(loss='categorical_crossentropy',
              optimizer=optimizer,
              metrics=['accuracy'])

if not data_augmentation:
    print('Not using data augmentation.')
    model.fit(x_train, y_train,
              batch_size=batch_size,
              epochs=epochs,
              validation_data=(x_test, y_test),
              shuffle=True)
else:
    print('Using real-time data augmentation.')
    # This will do preprocessing and realtime data augmentation:
    datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        zca_epsilon=1e-06,  # epsilon for ZCA whitening
        rotation_range=0,  # randomly rotate images in the range (degrees, 0 to 180)
        # randomly shift images horizontally (fraction of total width)
        width_shift_range=0.1,
        # randomly shift images vertically (fraction of total height)
        height_shift_range=0.1,
        shear_range=0.,  # set range for random shear
        zoom_range=0.,  # set range for random zoom
        channel_shift_range=0.,  # set range for random channel shifts
        # set mode for filling points outside the input boundaries
        fill_mode='nearest',
        cval=0.,  # value used for fill_mode = "constant"
        horizontal_flip=True,  # randomly flip images
        vertical_flip=False,  # randomly flip images
        # set rescaling factor (applied before any other transformation)
        rescale=None,
        # set function that will be applied on each input
        preprocessing_function=None,
        # image data format, either "channels_first" or "channels_last"
        data_format=None,
        # fraction of images reserved for validation (strictly between 0 and 1)
        validation_split=0.0)

    # Compute quantities required for feature-wise normalization
    # (std, mean, and principal components if ZCA whitening is applied).
    datagen.fit(x_train)

    # Fit the model on the batches generated by datagen.flow().
    model.fit_generator(datagen.flow(x_train, y_train,
                                     batch_size=batch_size),
                        epochs=epochs,
                        validation_data=(x_test, y_test),
                        workers=4)

scores = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```
